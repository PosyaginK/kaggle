## First: change NaN Age
First I replaced the missing data with a mod. But in this case, there are too many people whose age is 28 years old.
This is not very realistic, so I decided to fill in the blanks proportionally.

### Let's write the functions that we need

```python
def get_percent(df, c): # We calculate the percentage, which is the number of people of a certain age from the total number of people whose age is known
    per_list = []
    for i in df.Age:
        per = ((i * 100) / c).round(3)
        per_list.append(per)
        
    return per_list

def from_per_to_pas(df, l, c): # We translate the percentage obtained above to the number of people whose age is not known
    j = 0 
    for i in df.index:
        count = (l[j] * c) / 100
        if count < 1:
            count = 1
        df['Age'][i] = count
        j += 1    
        
    return df

def make_list(df): # Create a list with ages, from which we will fill in the missing values
    list_g = []
    for i in df.index:
        g = [i] * df['Age'][i]
        if df['Age'][i] < 1:
            g = [i] * 1
        list_g.append(g)
    
    for i in range(len(list_g) - 1):
        list_g[0] += list_g[i + 1]
        
    return np.array(list_g[0])

def insert_to_df(nan, df, l): # Insert the obtained values into the train
    j = 0
    for i in nan.index:
        df.Age[i] = l[j]
        j += 1
    return df
```

```python
def get_age(df):
    
    ages = df['Age'].value_counts() # Total different ages in the dataframe with the number of passengers with that age
    
    count_pas = df['PassengerId'].value_counts().sum() # Total passengers
    count_age = ages.sum() # Number of people whose age is known
    count_nan_age = count_pas - count_age # The number of people whose age is not known
    
    # We calculate what age occurs most often:
    med = df['Age'].mode()[0]
    
    ages = pd.DataFrame(ages) # Bringing a series to a dataframe
    
    # We calculate the percentage, which is the number of people of a certain age from the total number of people whose age is known
    per_list = get_percent(ages, count_age)
        
    # We translate the percentage obtained above to the number of people whose age is not known
    ages = from_per_to_pas(ages, per_list, count_nan_age)
    
    # Create a list with ages, from which we will fill in the missing values
    list_g = make_list(ages)
    
    # If there are fewer elements in our list than the missing values in the dataframe, then the remaining values are replaced with the mode
    if len(list_g) < count_nan_age:
        div = count_nan_age - len(list_g)
        add_el = [med]*div
        list_g = np.concatenate((list_g[:], add_el))
    elif len(list_g) > count_nan_age:
        list_g = l[:count_nan_age]
        
    nan_age = train[pd.isnull(train['Age'])] # Create a DF where age is unknown
    
    df = insert_to_df(nan_age, df, list_g) # Insert the obtained values into the train
    
    return df
```
